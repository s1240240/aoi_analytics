```python
#線形回帰モデル
from sklearn.linear_model import LinearRegression
```

```python
import mglearn
```

```python
x,y=mglearn.datasets.make_wave(n_samples=60)
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42)
#学習データとテストデータに分割を行う。その前に擬似乱数を用いて、データセットをシャッフルする。
```

```python
lr=LinearRegression().fit(x_train,y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
print("train set score: {:.2f}".format(lr.score(x_train,y_train)))
print("test set score:{:.2f}".format(lr.score(x_test,y_test)))
#学習データの決定係数
#テストデータの決定係数
```

```python
x,y=mglearn.datasets.load_extended_boston()
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)
```

```python
lr=LinearRegression().fit(x_train,y_train)
```

```python
print("train set score: {:.2f}".format(lr.score(x_train,y_train)))
print("test set score:{:.2f}".format(lr.score(x_test,y_test)))
#過剰適合が起きている
```

```python
#リッジ回帰
from sklearn.linear_model import Ridge

ridge=Ridge().fit(x_train,y_train)
print("train set score: {:.2f}".format(ridge.score(x_train,y_train)))
print("test set score:{:.2f}".format(ridge.score(x_test,y_test)))
```

```python
from sklearn.linear_model import Lasso
```

```python
lasso=Lasso().fit(x_train,y_train)
```

```python
import numpy as np
print("train set score: {:.2f}".format(lasso.score(x_train,y_train)))
print("test set score:{:.2f}".format(lasso.score(x_test,y_test)))
print("number of features used:{}".format(np.sum(lasso.coef_!=0)))
```

```python
lasso1=Lasso(alpha=0.01,max_iter=10000).fit(x_train,y_train)
```

```python
print("train set score: {:.2f}".format(lasso1.score(x_train,y_train)))
print("test set score:{:.2f}".format(lasso1.score(x_test,y_test)))
print("number of features used:{}".format(np.sum(lasso1.coef_!=0)))
```

```python
#カテゴリデータのダミー化
#p211
#1線形回帰で行なったダミー変数を整形する前のデータをcsvfileに変換し、machine_learningのディレクトリに置く
#2それを使い、dtypesを使い、型を確認し、ダミー化したいカラムを文字列(object型)に変換する
#3get_dummiesを使ってdataframe全てを対象にダミー化する(数値データはダミー化されない)
#4dataframeをターゲットと特徴量に分ける 
#5sklearn のいくつかの回帰モデルを使って学習させる
#alphaの値を調節して良い結果が出ればそれを報告する
```

```python
#np.linespace()は等差数列を生成する関数
#start=数列の始点
#end=数列の終点
#num=生成する配列の要素数
#endpoint=生成する配列においてTrueの場合stopの要素を含むFalseの場合stopの要素を含まない
#np.reshape
#配列の形状を変える
#引数に(n,-1)とするとn*mの2次元配列となる
#n*m=配列の要素数　となるようなmを決める
x,y=mglearn.datasets.make_wave(n_samples=100)
line=np.linspace(-3,3,1000,endpoint=False).reshape(-1,1)
```

```python
x
```

```python
line
```

```python
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt
reg=DecisionTreeRegressor(min_samples_split=3).fit(x,y)
plt.plot(line,reg.predict(line),label="decision tree")
reg=LinearRegression().fit(x,y)
plt.plot(line,reg.predict(line),label="linear regression")
plt.plot(x[:,0],y,"o",c="k")
plt.ylabel("Regression output")
plt.xlabel("Input feature")
plt.legend(loc="best")
```

```python
bins=np.linspace(-3,3,11)
print(bins)
```

```python
which_bin=np.digitize(x,bins=bins)
```

```python
print("\nData point:\n",x[:5])
print("\nBin membership for data points:\n",which_bin[:5])
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
#OneHotEncoderで変換する
encoder=OneHotEncoder(sparse=False)
#encoder.fitでwhich_binに現れる整数値のバリエーションを確認
encoder.fit(which_bin)
#transformでワンホットエンコーディングを行う
x_binned=encoder.transform(which_bin)
```

```python
print(x_binned)
```

```python
line_binned=encoder.transform(np.digitize(line,bins=bins))
reg=LinearRegression().fit(x_binned,y)
plt.plot(line,reg.predict(line_binned),label="linear regression binned")
reg=DecisionTreeRegressor(min_samples_split=3).fit(x_binned,y)
plt.plot(line,reg.predict(line_binned),label="decision tree binned")
plt.plot(x[:,0],y,"o",c="k")
plt.vlines(bins,-3,3,linewidth=1,alpha=.2)
plt.legend(loc="best")
plt.ylabel("Regression output")
plt.xlabel("input feature")
```
