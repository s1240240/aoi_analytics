```python
import statsmodels.api as sm
import matplotlib.pyplot as plt
res=sm.tsa.seasonal_decompose(data2017,freq=168)#周期を１週間に調整
original=data2017
trend=res.trend
seasonal=res.seasonal
residual=res.resid
plt.figure(figsize=(8,8))
```

```python
plt.subplot(411)
plt.plot(original)
plt.subplot(412)
plt.plot(trend)
plt.ylabel("Trend")
plt.subplot(413)
plt.plot(seasonal)
plt.ylabel("Seasonal")
plt.subplot(414)
plt.plot(residual)
plt.ylabel("residual")
plt.tight_layout()
```

```python
sum_three_data=trend+seasonal+residual
plt.figure(figsize=(8,4))
plt.plot(original,label="original")
plt.plot(sum_three_data,label="trend+seasonal+residual",linestyle="--")
plt.legend(loc="best")
```

```python
fig = plt.figure(figsize=(16, 4))
ax1 = fig.add_subplot(211)
sm.graphics.tsa.plot_acf(data2017, lags=168, ax=ax1)
fig = plt.figure(figsize=(16, 4))
ax2 = fig.add_subplot(212)
sm.graphics.tsa.plot_pacf(data2017, lags=168, ax=ax2) #PACF計算とグラフ自動作成
plt.tight_layout()
```

```python
sarimax_train=sm.tsa.SARIMAX(data2017,order=(2,1,3),
               seasonal_order=(0,1,1,13),
               enforce_stationarity=False,
               enforce_invertiblity=False
                            ).fit()
```

```python
sarimax_train_pred=sarimax_train.predict("2017-01-01 09:00:00","2017-12-31 23:00:00")
plt.plot(data2017,c="r",label="actual")
plt.plot(sarimax_train_pred,c="b",label="model_pred",alpha=0.7)
plt.legend(loc="best")
```

```python
plt.figure(figsize=(12,4))
predict_12month=sarimax_train.get_prediction(start="2017-12-01 09:00:00",end="2017-12-31 23:00:00")
predict_12month_ci=predict_12month.conf_int(alpha=0.05)#信頼区間のデータ取得
plt.plot(data2017.query('"2017-12-01 09:00:00"<=date<="2017-12-31 23:00:00"'),label="actual",c="r")
plt.plot(predict_12month.predicted_mean,c="b",label="model_pre",alpha=0.7)
plt.fill_between(predict_12month_ci.index,predict_12month_ci.iloc[:,0],predict_12month_ci.iloc[:,1]\
                ,color="g",alpha=0.2)
plt.legend(loc="upper left")
```

```python
import warnings
warnings.filterwarnings('ignore') # 警告非表示（収束：ConvergenceWarning）

# パラメータ範囲
# order(p, d, q)
min_p = 1; max_p = 3 # min_pは1以上を指定しないとエラー
min_d = 0; max_d = 1
min_q = 0; max_q = 3 

# seasonal_order(sp, sd, sq)
min_sp = 0; max_sp = 1
min_sd = 0; max_sd = 1
min_sq = 0; max_sq = 1

test_pattern = (max_p - min_p +1)*(max_q - min_q + 1)*(max_d - min_d + 1)*(max_sp - min_sp + 1)*(max_sq - min_sq + 1)*(max_sd - min_sd + 1)
print("pattern:", test_pattern)

sfq = 12 # seasonal_order周期パラメータ
ts = data2017 # 時系列データ
test_results = pd.DataFrame(index=range(test_pattern), columns=["model_parameters", "aic"])
num = 0
for p in range(min_p, max_p + 1):
    for d in range(min_d, max_d + 1):
        for q in range(min_q, max_q + 1):
            for sp in range(min_sp, max_sp + 1):
                for sd in range(min_sd, max_sd + 1):
                    for sq in range(min_sq, max_sq + 1):
                        sarima = sm.tsa.SARIMAX(
                            ts, order=(p, d, q), 
                            seasonal_order=(sp, sd, sq, sfq), 
                            enforce_stationarity = False, 
                            enforce_invertibility = False
                        ).fit()
                        test_results.iloc[num]["model_parameters"] = "order=(" + str(p) + ","+ str(d) + ","+ str(q) + "), seasonal_order=("+ str(sp) + ","+ str(sd) + "," + str(sq) + ")"
                        test_results.iloc[num]["aic"] = sarima.aic
                        print(num,'/', test_pattern-1, test_results.iloc[num]["model_parameters"],  test_results.iloc[num]["aic"] )
                        num = num + 1

# 結果（最小AiC）
print("best[aic] parameter ********")
print(test_results[test_results.aic == min(test_results.aic)])
```
